````
import cv2
def xmin(i):
  try:
   img = cv2.imread(i)
   hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
   lower = (0,255,255)
   upper = (150,255,255)
   thresh = cv2.inRange(hsv, lower, upper)
   result = np.zeros_like(img)
   contours = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
   contours = contours[0] if len(contours) == 2 else contours[1]
   cntr = contours[0]
   cv2.waitKey(0)
   cv2.destroyAllWindows()
   w=pd.DataFrame(cntr.reshape(-1,1),columns=['shapes'])
   k=w.iloc[::2]
   j=w.iloc[1::2]
   x=min(k.shapes)
   return x
  except IndexError:
     pass

def xmax(i):
  try:
   img = cv2.imread(i)
   hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
   lower = (0,255,255)
   upper = (150,255,255)
   thresh = cv2.inRange(hsv, lower, upper);
   result = np.zeros_like(img)
   contours = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
   contours = contours[0] if len(contours) == 2 else contours[1]

   cntr = contours[0]
   cv2.waitKey(0)
   cv2.destroyAllWindows()
   w=pd.DataFrame(cntr.reshape(-1,1),columns=['shapes'])
   k=w.iloc[::2]
   j=w.iloc[1::2]
   x=max(k.shapes)
   return x
  except IndexError:
     pass
def ymin(i):
  try:
   img = cv2.imread(i)
   hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
   lower = (0,255,255)
   upper = (150,255,255)
   thresh = cv2.inRange(hsv, lower, upper);
   result = np.zeros_like(img)
   contours = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
   contours = contours[0] if len(contours) == 2 else contours[1]
   cntr = contours[0]
   cv2.waitKey(0)
   cv2.destroyAllWindows()
   w=pd.DataFrame(cntr.reshape(-1,1),columns=['shapes'])
   k=w.iloc[::2]
   j=w.iloc[1::2]
   x=min(j.shapes)
   return x
  except IndexError:
     pass

def ymax(i):
  try:
   img = cv2.imread(i)
   hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
   lower = (0,255,255)
   upper = (150,255,255)
   thresh = cv2.inRange(hsv, lower, upper);
   result = np.zeros_like(img)
   contours = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
   contours = contours[0] if len(contours) == 2 else contours[1]
   cntr = contours[0]
   cv2.waitKey(0)
   cv2.destroyAllWindows()
   w=pd.DataFrame(cntr.reshape(-1,1),columns=['shapes'])
   k=w.iloc[::2]
   j=w.iloc[1::2]
   x=max(j.shapes)
   return x
  except IndexError:
     pass

def width(i):
  im = Image.open(i)
  width, height = im.size
  return width

def he(i):
  im = Image.open(i)
  width, he = im.size
  return he
````

````
#center of rectange
def centerx (i):
  coord = (trainrec1.xmin[i],trainrec1.ymin[i],trainrec1.xmax[i],trainrec1.ymax[i])
  centerCoord = (coord[0]+(coord[2]/2), coord[1]+(coord[3]/2))
  return centerCoord[0]

def centery (i):
  coord = (trainrec1.xmin[i],trainrec1.ymin[i],trainrec1.xmax[i],trainrec1.ymax[i])
  centerCoord = (coord[0]+(coord[2]/2), coord[1]+(coord[3]/2))
  return centerCoord[1]

def widthbox (i):
   w = (trainrec1.xmax[i]-trainrec1.xmin[i])
   return w


def heightbox (i):
   w = (trainrec1.ymax[i]-trainrec1.ymin[i])
   return w


trainrec1['x']=0
trainrec1['y']=0
trainrec1['d'] = trainrec1.index
trainrec1['x']=trainrec1['d'].apply(lambda x:(centerx(x)))
trainrec1['y']=trainrec1['d'].apply(lambda x:(centery(x)))
trainrec1['widhtbox']=trainrec1['d'].apply(lambda x:(centery(x)))
trainrec1['heightbox']=trainrec1['d'].apply(lambda x:(centery(x)))

del trainrec1['d']

trainrec1.dropna()
````

````
def js_reply_to_image(js_reply):
    """
    input: 
          js_reply: JavaScript object, contain image from webcam
    output: 
          image_array: image array RGB size 512 x 512 from webcam
    """
    jpeg_bytes = base64.b64decode(js_reply['img'].split(',')[1])
    image_PIL = Image.open(io.BytesIO(jpeg_bytes))
    image_array = np.array(image_PIL)

    return image_array
````
